```python
# Initiate the algoritnm
# Use clustering techniques to segment the customers
from sklearn.cluster import KMeans


# Use PCA to reduce the dimensionality of the data
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
segm_pca = pca.fit_transform(segm[['Customer Age', 'Quantity','Cost','Customer Gender_M']])
```

```python
from sklearn.preprocessing import StandardScaler

# Create an instance of the StandardScaler class
scaler = StandardScaler()

# Fit the scaler to your data
scaler.fit(segm_pca)

# Use the transform method to standardize the data
segm_pca = scaler.transform(segm_pca)
```

```python
#using the elbow method to find out the optimal number of #clusters. 
#Loop through to fit the k-means algorithm to the data

intertia_scores =[]
for i in range (1, 11):
    kmeans = KMeans(n_clusters = i, init ='k-means++', max_iter=300,  n_init=10,random_state=0 )
    kmeans.fit(segm_pca)
    intertia_scores.append(kmeans.inertia_)
```

```python
plt.plot(range (1,11), intertia_scores)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of clusters')
plt.ylabel('intertia_scores')
plt.show()     
```

```python
#According to the Elbow graph the clusters number is 3. 
#Applying k-means algorithm to the segm dataset.

kmeans = KMeans( n_clusters = 3, init ='k-means++', max_iter=300, n_init=10,random_state=0 ).fit(segm_pca)
centroids = kmeans.cluster_centers_
print(centroids)

# Assign the segment labels to the customers
segm['Segment']  = kmeans.labels_
segm.head()
```

```python
# Explore the segments using visualizations
sns.scatterplot(x=segm_pca[:,0], y=segm_pca[:,1], hue='Segment', data=segm)
plt.show()
```

```python
# Break the result down by gender 
print(pd.crosstab(segm['Segment'], segm['Customer Gender_M']))
print(pd.crosstab(segm['Segment'], segm['Customer Gender_M'], normalize = 'index'))
```
